# Pandas 시각화

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
mpl.rcParams['axes.unicode_minus']=False
mpl.rcParams['font.family'] = 'Malgun Gothic'
```

### 1.Line graph

```python
np.random.seed(2022)
df = pd.DataFrame(np.random.randn(100, 3),
                    index=pd.date_range('2022-1-1',periods=100),
                    columns=['A','B','C']).cumsum()
df.head()
                
```

```python
df.plot()
plt.title('Pandas의 plot method 사용예')
plt.xlabel('일자'), plt.ylabel('데이터')
plt.show()
```

## 2. Bar chart 

```python
import seaborn as sns
iris = sns.load_dataset('iris')

plt.figure(figsize=(12,6))
iris.sepal_length[:50].plot(kind='bar',rot=45)  #bar
plt.title('setosa 품종의 sepal_length',fontsize=20)
plt.show()
```

```python
iris.sepal_length[:20].plot(kind='barh',rot=45)     #barh
plt.title('setosa 품종의 sepal_length',fontsize=20)
plt.show()
```

```python
df2 =iris.groupby('species').mean()
df2
```

```python
df2.plot.bar(rot=0)
plt.title('Iris 각 품종의 Feature별 평균',fontsize=20)
plt.show()
```

```python
#범례의 모양 조정
df2.columns.name = 'Feature'
df2.plot.bar(rot=0)
plt.title('Iris 각 품종의 Feature별 평균',fontsize=20)
plt.ylim(0,9)
plt.show()
```

## 3. Pie chart

```python
titanic = sns.load_dataset('titanic')
s = titanic['class'].value_counts()
s
```

```python
s.plot.pie(autopct='%1f%%')
plt.title('선실별 승객수')
plt.show()
```

## 4. Box plot

```python
iris[iris.species == 'setosa'].plot.box(figsize=(6,8))
plt.title('Setosa 품종의 각 피쳐별 box plot')
plt.show()
```
