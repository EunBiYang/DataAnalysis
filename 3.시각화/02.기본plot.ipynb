# Matplotlib - 기본 plot

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
mpl.rcParams['axes.unicode_minus']=False
mpl.rcParams['font.family'] = 'Malgun Gothic'
```

```python
plt.plot([0,1,4,9])     #x,y (x가 생략되면 0,1,2,3,4,....을 의미)
plt.title('라인 플롯')
plt.grid()
plt.show()
```

-  Line style

```python
plt.plot([0,1,2,3],[0,1,4,9],'ro-')     #' ro-' :color, maker,linestyle
plt.title('Red-Circle-직선')
plt.grid()
plt.show()
```

```python
plt.plot([0,1,2,3],[0,1,4,9],'g*--')     #' ro-' :color, maker,linestyle
plt.title('Green-Star-점선',fontsize=20)
plt.grid()
plt.show()
```

```python
plt.plot([0,1,2,3],[0,1,4,9], 'ms--')     # color, marker, linestyle
plt.plot([8,4,2,1], 'g^:')
plt.title('Magenta-Square-Dashed, Green-삼각형-점선',fontsize=20)
plt.grid()
plt.show()
```

```python
plt.plot([0,1,2,3],[0,1,4,9], 'ms--', linewidth=3, markersize=10)
plt.plot([8,4,2,1], 'g^:', lw=5, ms=20)
plt.title('Magenta-Square-Dashed, Green-삼각형-점선')
plt.grid()
plt.show()
```

```python
# color   :  blue, green, red, cyan, magenta, yellow, black(k), white
# marker  :  circle(o), 삼각형(^), 역삼각형(v), 사각형(s), 별(*), 플러스(+)
# linestyle  : 직선(-), dashed(--), dash-dot(-.), dotted(:)
```

- 그림의 범위

```python
plt.plot([0,1,2,3],[0,1,4,9],'ro-')     #' ro-' :color, maker,linestyle
plt.title('X축 Y축 범위지정',fontsize=20)
plt.xlim(-1,4); plt.ylim(-1,10)
plt.grid(); plt.show()
```

- 곡선 그림

```python
X = np.arange(0, 2*np.pi, 0.01)     #(,  ,간격)
Y = np.sin(X)
plt.plot(X,Y)
plt.title('Sine 그래프',fontsize=20)
plt.grid();plt.show()
```

```python
X = np.arange(0, 2*np.pi, 0.01)     #(,  ,간격)
Ysin, Ycos = np.sin(X), np.cos(X)
plt.plot(X,Ysin), plt.plot(X,Ycos)
plt.title('Sine 그래프',fontsize=20)
plt.grid();plt.show()
```

- Tick 설정

```python
X = np.arange(0, 2*np.pi, 0.01)     #(,  ,간격)
Ysin, Ycos = np.sin(X), np.cos(X)
plt.plot(X,Ysin), plt.plot(X,Ycos)
plt.title('Tick 설정',fontsize=20)
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],    #tick
            [r'$0$', r'$\pi/2$', r'$\pi$', r'$3\pi/2$', r'$2\pi$'])  #화면에 이쁘게 표시되게 하는거
plt.yticks([-1, 0, 1])
plt.grid();plt.show()
```

- 범례(Legend)

```python
X = np.arange(0, 2*np.pi, 0.01)     #(,  ,간격)
Ysin, Ycos = np.sin(X), np.cos(X)
plt.plot(X,Ysin,label='sine'), plt.plot(X,Ycos,label='cosine')
plt.title('범례 표시',fontsize=20)
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
            [r'$0$', r'$\pi/2$', r'$\pi$', r'$3\pi/2$', r'$2\pi$'])  #LaTex
plt.yticks([-1, 0, 1])
plt.legend(loc=0);plt.grid();plt.show()      #레전드 꼭 앞에. 뒤에 있으니 오류났엇음!!
#레전드 괄호안에 loc=로 위치설정 가능, 입력숫자는 교재에서 확인.생략도 가능
```

- X축 Y축 Label

```python
plt.plot(X,Ysin,label='sine'), plt.plot(X,Ycos,label='cosine')
plt.title('X축 Y축 Label표시',fontsize=20)
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
            [r'$0$', r'$\pi/2$', r'$\pi$', r'$3\pi/2$', r'$2\pi$'])  #LaTex
plt.yticks([-1, 0, 1])
plt.xlabel('시간'), plt.ylabel('진폭')
plt.legend(loc=0);plt.grid();plt.show() 
```

- 그림의 크기

```python
plt.figure(figsize=(10,6))  #(가로,세로)
plt.plot(X,Ysin,label='sine'), plt.plot(X,Ycos,label='cosine')
plt.title('그림크기지정(10,6)',fontsize=20)
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
            [r'$0$', r'$\pi/2$', r'$\pi$', r'$3\pi/2$', r'$2\pi$'])  #LaTex
plt.yticks([-1, 0, 1])
plt.xlabel('시간'), plt.ylabel('진폭')
plt.legend(loc=0);plt.grid();plt.show() 
```

- 그림 저장

```python
plt.figure(figsize=(10,6))
plt.plot(X, Ysin, label='sine'), plt.plot(X, Ycos, label='cosine')
plt.title('그림 저장', fontsize=15)
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
           [r'$0$', r'$\pi/2$', r'$\pi$', r'$3\pi/2$', r'$2\pi$'])    # LaTeX
plt.yticks([-1, 0, 1])
plt.xlabel('시간'), plt.ylabel('진폭')
plt.legend(loc=0); plt.grid()
plt.savefig('삼각함수.png')
```
