- Titanic


```python
import numpy as np
import pandas as pd
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')
```

```python
titanic = sns.load_dataset('Titanic')
```

<pre>
a. 타이타닉호의 승객에 대해 나이와 성별에 의한 카테고리 열인 category1 열을 만드시오. 
    category1 카테고리는 다음과 같이 정의됨
    1) 20살이 넘으면 성별을 그대로 사용한다.
    2) 20살 미만이면 성별에 관계없이 “child”라고 한다.

b. 타이타닉호의 승객 중 나이를 명시하지 않은 고객은 나이를, 명시한 고객의 평균 나이값이 되도록 titanic 데이터프레임을 고치시오.

c. 성별, 선실(class)별, 출발지(embark_town)별 생존율을 구하시오.

d. 타이타닉호 승객을 ‘미성년자’, ‘청년’, ‘중년’, ‘장년’, ‘노년’ 나이 그룹으로 나누고, 각 그룹별 생존율을 구하시오.
    bins = [1, 20, 30, 50, 70, 100]
    labels = ["미성년자", "청년", "중년", "장년", "노년"]

e. qcut 명령으로 세 개의 나이 그룹을 만들고, 나이 그룹별 남녀 성비와 생존율을 구하시오.
<pre>

```python
# 문제 a
titanic['category1'] = titanic.age.apply(lambda x: 'child' if x <20  else x)
titanic.tail()
```

```python
#강사님답
titanic['category2'] = titanic.apply(lambda r : r.sex if r.age >= 20 else'child', axis=1)
titanic.tail()
```

```python
# 문제 b
titanic.fillna(titanic.age.mean()).round(1)
```

```python
#강사님 답
titanic.age.fillna(titanic.age.mean(), inplace=True)
titanic.tail()
```

```python
#c. 성별, 선실(class)별, 출발지(embark_town)별 생존율을 구하시오.
titanic.groupby(['sex','class','embark_town'])[['survived']].mean()
```

```python
titanic.pivot_table('survived','sex','class')
```

```python
# d. 타이타닉호 승객을 ‘미성년자’, ‘청년’, ‘중년’, ‘장년’, ‘노년’ 나이 그룹으로 나누고, 각 그룹별 생존율을 구하시오.
age = titanic['age']
bins = [1, 20, 30, 50, 70, 100]
labels = ["미성년자", "청년", "중년", "장년", "노년"]

titanic['age_cat'] = pd.cut(age, bins, labels=labels)
titanic.groupby(['age_cat'])[['survived']].mean()
```

```python
#피봇 테이블의 형태로
titanic.pivot_table('survived', "age_cat")
```

```python
# e. qcut 명령으로 세 개의 나이 그룹을 만들고, 나이 그룹별 남녀 성비와 생존율을 구하시오.
titanic['age_group'] = pd.qcut(titanic.age, 3, labels=('A1','A2','A3'))
titanic.groupby(['age_group','sex'])[['survived']].mean()
```

```python
titanic.groupby('age_group')[['survived']].mean()
```

```python
titanic.tail()
```

- MPG 

```python
mpg = sns.load_dataset('mpg')
```

```python
#a 배기량
mpg['hp_per_cc'] = mpg.displacement / mpg.horsepower
```

```python
mpg.tail()
```

```python
# b문제
mpg['manufacturer'] = mpg.name.apply(lambda x : x.split()[0])
mpg['model'] = mpg.name.apply(lambda x : ''.join(x.split()[1:]))
mpg.drop(columns=['name'], inplace=True)
mpg.head()
```

```python
#str 조작 매소드 이용
mpg['manufature'] = mpg.name.str.split().str[0]
mpg['model'] = mpg.name.str.split().str[1:].str.join('')
mpg.drop(columns=['name'], inplace=True)
mpg.tail()
```

```python
# c. 엔진의 실린더(cylinders) 갯수별 연비(mpg)의 평균을 구하시오.
mpg.groupby(['cylinders'])[['mpg']].mean()
```

```python
mpg.pivot_table('mpg','cylinders')
```

```python
mpg.head()
```

```python
# d. 생산지(origin)별 배기량 대비 마력(hp_per_cc)의 평균을 구하시오

mpg.groupby('origin')[['hp_per_cc']].mean()
```

```python
mpg.pivot_table('hp_per_cc', 'origin')
```

```python
# e. 모델이 5개 이상인 제조사에 대하여 연비(mpg)의 평균이 가장 좋은 제조사 Top 5를 구하시오.
mpg.groupby(['manufacturer','model'])[['mpg']].mean()
```

```python
mpg.head()
```

```python
#강사님 답
mpg.manufacturer.value_counts().head()
```

```python
mpg.manufacturer.value_counts().tail()
```

```python
#s[s >= 5] 이게 필터링!!  = 모델이 5개 이상인 제조사.
s = mpg.manufacturer.value_counts()
s[s >= 5]
```

```python
#위의 시리즈에서 행열이니깐 인덱스!!
s[s >= 5].index
```

```python
#big5 = list(s[s >= 5].index)
big5 = s[s >= 5].index.tolist()
print(big5)
```

```python
#mpg 데이터에서 big5에 속하는 것만 필터링
#isin 메소드.
mpg.manufacturer.isin(big5).head()
```

```python
#데이터 프레임 형식으로 변환하면서 위에서 True가 나오 값만 정리.
mpg[mpg.manufacturer.isin(big5)].head()
```

```python
big5_mpg = mpg[mpg.manufacturer.isin(big5)]
big5_mpg.pivot_table('mpg','manufacturer')
```

```python
big5_mpg.pivot_table('mpg','manufacturer').sort_values(by='mpg',ascending =False).head()
```

```python
s = mpg.manufacturer.value_counts()
big5 = s[s >= 5].index.tolist()
big5_mpg = mpg[mpg.manufacturer.isin(big5)]
big5_mpg.pivot_table('mpg','manufacturer').sort_values(by='mpg',ascending =False).head()
```
