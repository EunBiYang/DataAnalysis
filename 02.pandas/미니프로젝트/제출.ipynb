```python
import numpy as np
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

## 1)CCTV 데이터 정리하기

```python
#데이터 파일 읽기
cctv = pd.read_csv('서울시CCTV설치운영현황(자치구)_년도별_211231기준.csv', encoding='EUC-KR',skiprows=1,thousands=',')
cctv.head()
```

```python
#Null data 확인하기
cctv.isna().sum()
```

```python
#Null값 0으로 대체하기
cctv.fillna('0',inplace=True)
```

```python
#2021년을 제외한 최근 3개년간의 CCTV 증가율 구하기
a = cctv.loc[:,['2012년','2013년','2014년','2015년','2016년','2017년']].sum(axis=1)
b = cctv.loc[:,['2018년','2019년','2020년']].sum(axis=1)*100
cctv['최근증가율'] = b/a
cctv.head()
```

```python
#구분, 총계, 최근증가율 커럼만 추출하고 컬럼명을 구별, cctv댓수로 변경하기
df = cctv[['구분','총계','최근증가율']]
df.rename(columns={'구분':'구별','총계':'cctv댓수'}, inplace=True)
df.head()
```

```python
#구 이름에서 공백 지우기
df['구별'] = df['구별'].str.replace(' ', '')
df.head()
```

```python
#첫번째 행(계) 지우기
df.drop(0, inplace=True)
df.head()
```

## 2)인구 데이터 정리하기

```python
#데이터 파일 읽기
pop = pd.read_table('서울시인구.txt',skiprows=2,thousands=',')
pop.head()
```

```python
#Null data 있는지 확인하기
pop.isna().sum()
```

```python
#'자치구','계','계.1','계.2','65세이상고령자‘ 컬럼만 추출하고
# 이름을 '구별','인구수','내국인','외국인','고령자'로 변경
df2 = pop[['자치구','계','계.1','계.2','65세이상고령자']]
df2.columns = ['구별','인구수','내국인','외국인','고령자']
df2.head()
```

```python
#첫번째 행 지우기
df2.drop(0,inplace=True)
df2.head()
```

```python
#'외국인비율'과'고령자비율'컬럼만들기
df2['외국인비율'] = ''
df2['고령자비율'] = ''
df2.head()
```

## CCTV데이터와 인구 데이터 합치고 저장하기

```python
df.head()
```

```python
df2.head()
```

```python
# 데이터 합치기
cctv_pop = pd.merge(df,df2)
cctv_pop.head()
```

```python
#‘구별’ 컬럼을 인덱스로 만들기
cctv_pop.set_index('구별', inplace=True)
cctv_pop.head()
```

```python
# 합친 데이터를 cctv_pop.csv로 저장하기
cctv_pop.to_csv('cctv_pop.csv')
```
