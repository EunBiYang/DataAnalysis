# Matplotlib - 다양한 플롯

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
mpl.rcParams['axes.unicode_minus']=False
mpl.rcParams['font.family'] = 'Malgun Gothic'
```

### 1. 산점도(Sctter Plot)
    - x와 y의 관계를 눈으로 확인하고자 할때

```python
x,y = np.random.randn(100), np.random.randn(100)
plt.scatter(x,y)
plt.title('산점도', fontsize=20)
plt.show()
```

```python
colors = np.random.rand(100)        #랜덤한 색상을 주는것
plt.scatter(x,y, c=colors, alpha=0.5)       #색과 투명도
plt.title('색상을 준 산점도',fontsize=20)
plt.show()
```

## Histogram

```python
s = np.random.randn(200)
plt.hist(s)
plt.title('Histogram',fontsize=20)
plt.show()
```

```python
s = np.random.randn(2000000)
plt.hist(s, bins=20)                #2만개를 20개로 나눠서
plt.title('Histogram',fontsize=20)
plt.show()
```

##  4. Box plot

```python
#기초통계량 확인 할때 사용된다.
```

```python
#setosa 품종의 sepal_witdh
iris = sns.load_dataset('iris')
setosa = iris[iris.species=='setosa']
```

```python
plt.boxplot(setosa.sepal_width)
plt.title('Setosa 품종, sepal_width',fontsize=20)
plt.show()
```

```python
#여러개를 그리고 싶을때
plt.boxplot([setosa.sepal_width,setosa.sepal_length],labels=iris.columns[:2])
plt.title('Setosa 품종, 꽃받침의 길이,너비',fontsize=20)
plt.show()
```

## 5.파이(Pie)차트

```python
ratio = [34, 32, 16, 18]
labels = ['감', '귤', '배', '밤']
plt.pie(ratio, labels=labels, autopct='%.1f%%')
plt.title('Pie chat',fontsize=20)
plt.show()
```

```python
#색상 변경해보기
colors = ['yellowgreen','gold','lightskyblue','lightcoral']
explode = (0,0,0.1,0)
plt.pie(ratio, labels=labels, explode=explode,colors=colors, autopct='%.1f%%')
plt.title('Pie chat',fontsize=20)
plt.show()
```
